## 2. Basic Data Analysis Demo

Test basic data science capabilities:

````python
# Import common data science libraries
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    print("‚úÖ Data science libraries imported successfully")
    print(f"   - pandas: {pd.__version__}")
    print(f"   - numpy: {np.__version__}")
    print(f"   - matplotlib: {plt.matplotlib.__version__}")
    print(f"   - seaborn: {sns.__version__}")
except ImportError as e:
    print(f"‚ùå Error importing libraries: {e}")
    print("Install with: pip install pandas numpy matplotlib seaborn")
````

````python
# Create sample data
np.random.seed(42)
data = {
    'date': pd.date_range('2024-01-01', periods=100, freq='D'),
    'value': np.random.randn(100).cumsum() + 100,
    'category': np.random.choice(['A', 'B', 'C'], 100),
    'metric': np.random.randn(100) * 10 + 50
}

df = pd.DataFrame(data)
print("Sample dataset created:")
print(df.head())
print(f"\nDataset shape: {df.shape}")
````

````python
# Create a visualization
plt.figure(figsize=(12, 4))

# Time series plot
plt.subplot(1, 2, 1)
plt.plot(df['date'], df['value'])
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)

# Category distribution
plt.subplot(1, 2, 2)
df['category'].value_counts().plot(kind='bar')
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

print("üìä Visualization created successfully!")
````

## 3. AI Development Tools Integration

Test integration with AI development tools:

````python
# Check for AI/ML libraries
ai_libraries = {
    'requests': 'HTTP requests',
    'openai': 'OpenAI API client',
    'langchain': 'LangChain framework',
    'transformers': 'Hugging Face Transformers',
    'chromadb': 'Chroma vector database',
    'ollama': 'Ollama Python client'
}

print("AI/ML Libraries Status:")
print("=" * 30)

for lib, description in ai_libraries.items():
    try:
        module = __import__(lib)
        version = getattr(module, '__version__', 'unknown')
        print(f"‚úÖ {lib} ({description}): {version}")
    except ImportError:
        print(f"‚ùå {lib} ({description}): Not installed")

print("\nüí° Install missing libraries in your conda environment or via pipx")
````

````python
# Test local LLM connection (Ollama)
import subprocess
import json

def check_ollama_status():
    """Check if Ollama is running and what models are available"""
    try:
        # Check if Ollama is running
        result = subprocess.run(['ollama', 'list'], 
                              capture_output=True, text=True, timeout=5)
        
        if result.returncode == 0:
            print("ü§ñ Ollama Status: Running")
            print("Available models:")
            print(result.stdout)
            return True
        else:
            print("‚ö†Ô∏è  Ollama Status: Not running or error")
            print(result.stderr)
            return False
            
    except subprocess.TimeoutExpired:
        print("‚è±Ô∏è  Ollama Status: Timeout (may not be running)")
        return False
    except FileNotFoundError:
        print("‚ùå Ollama Status: Not installed")
        print("Install with: brew install ollama")
        return False

ollama_available = check_ollama_status()
````

````python
# Test a simple AI request (if libraries are available)
def test_ai_integration():
    """Test AI integration with available tools"""
    
    # Test with requests (simple HTTP call)
    try:
        import requests
        
        # Test if Ollama API is accessible
        if ollama_available:
            try:
                response = requests.get('http://localhost:11434/api/tags', timeout=2)
                if response.status_code == 200:
                    models = response.json()
                    print("üåê Ollama API accessible")
                    print(f"Models available via API: {len(models.get('models', []))}")
                else:
                    print("‚ö†Ô∏è  Ollama API not responding")
            except requests.exceptions.RequestException:
                print("‚ùå Cannot connect to Ollama API")
    
    except ImportError:
        print("‚ùå requests library not available")
    
    # Test langchain if available
    try:
        from langchain.llms import Ollama
        print("‚úÖ LangChain with Ollama support available")
    except ImportError:
        print("‚ùå LangChain not available")
    
    # Test vector database
    try:
        import chromadb
        print("‚úÖ ChromaDB vector database available")
    except ImportError:
        print("‚ùå ChromaDB not available")

test_ai_integration()
````
